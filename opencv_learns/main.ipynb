# 图像位运算


```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img1=cv.imread("images/img1.png")
logo=cv.imread("images/logo.png")

logo=cv.resize(logo,(200,200))

print(img1.shape)

plt.imshow(img1)
```

```python
plt.imshow(logo)
print(logo.shape)
```

```python

imgout=cv.bitwise_and(img1,logo,)
plt.imshow(imgout)
```
